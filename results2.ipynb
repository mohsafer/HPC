# Load dataset

```python
x_train, x_test, x_val, y_train, y_test, y_val = load_data("dataset", (0.7, 0.2, 0.1))
```

## Encode classes as a one-hot numeric array

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
enc = OneHotEncoder(categories='auto')
enc.fit(np.concatenate((y_train, y_test), axis=0).reshape(-1, 1))
y_train_enc = enc.transform(y_train.reshape(-1, 1)).toarray()
y_test_enc = enc.transform(y_test.reshape(-1, 1)).toarray()
y_true = np.argmax(y_test, axis=1)
```

# LSTM

```python
input_shape = (60, 23)  
num_classes = 2          
model = lstm.LSTMModel(input_shape=input_shape, num_classes=num_classes).model
history = model.fit(x_train, y_train_enc, batch_size=100, epochs=100, validation_data=(x_test, y_test_enc))

y_true = np.argmax(y_test_enc, axis=1)  # Convert one-hot encoded to class indices
y_pred = model.predict(x_test)
y_pred = np.argmax(y_pred, axis=1)  # Convert predictions to class indices

# Duration of training (example)
duration = 300  # Duration in seconds
# Call the plot function and calculate metrics
plot(history, model, y_true, y_pred, duration)

#from plot_model import plot_model
 
from tensorflow.keras.utils import plot_model

plot_model(model, to_file='lstmplot.png', show_dtype=True, show_shapes=True, show_layer_names=True, rankdir='TB', expand_nested=True, dpi=96)





####################################################### Precision-Recall Curve
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

# Predict probabilities
y_proba = model.predict(x_test).ravel()
precision, recall, _ = precision_recall_curve(y_test_enc.ravel(), y_proba)

# Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.')
plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
#plt.grid()
plt.show()

####################################################ROC
from sklearn.metrics import roc_curve, auc

# Predict probabilities
y_proba = model.predict(x_test).ravel()
fpr, tpr, _ = roc_curve(y_test_enc.ravel(), y_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(8, 6))
plt.hist(y_pred, bins=20, alpha=0.7, label='Predicted Probabilities')
plt.title('Histogram of Predicted Probabilities')
plt.xlabel('Probability')
plt.ylabel('Frequency')
plt.legend()
plt.show()

#######################################################

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np

# Predict on test data
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test_enc, axis=1)

# Compute confusion matrix
cm = confusion_matrix(y_true_classes, y_pred_classes)

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=range(num_classes))
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
##################################################################
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the function to plot
def func(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) - 0.5 * (x**2 - y**2)

# Create a grid of points
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

# Create the 3D surface
surface = go.Surface(z=Z, x=X, y=Y, colorscale='Viridis', opacity=0.8)

# Add contour projections
contours = go.Contour(z=Z, x=x, y=y, colorscale='Viridis', opacity=0.8)

# Add gradient descent path
path_x = [1.5, 1.2, 0.8, 0.4, 0.1]
path_y = [1.5, 1.0, 0.7, 0.3, 0.0]
path_z = [func(x, y) for x, y in zip(path_x, path_y)]
path = go.Scatter3d(x=path_x, y=path_y, z=path_z, mode='markers+lines', line=dict(color='black'))

# Combine all plots
fig = go.Figure(data=[surface, path])
fig.update_layout(scene=dict(zaxis=dict(title='Z-axis'),
                             xaxis=dict(title='X-axis'),
                             yaxis=dict(title='Y-axis')),
                  title='Interactive 3D Surface with Gradient Path')

fig.show()

```

```python
import matplotlib.pyplot as plt

from matplotlib.patches import FancyBboxPatch

# Update positions of boxes to include small gaps
layers_with_gap = [
    {"name": "Input Layer\n(60, 23)", "x": 0, "y": 0, "width": 1.5, "height": 0.5},
    {"name": "SimpleRNN_1\n(60, 50)", "x": 2.5, "y": 0, "width": 1.5, "height": 0.5},
    {"name": "SimpleRNN_2\n(50)", "x": 5, "y": 0, "width": 1.5, "height": 0.5},
    {"name": "Flatten\n(50)", "x": 7.5, "y": 0, "width": 1.5, "height": 0.5},
    {"name": "Dense\n(2)", "x": 10, "y": 0, "width": 1.5, "height": 0.5},
]

# Create the plot with updated positions
fig, ax = plt.subplots(figsize=(12, 3))

# Add boxes for each layer
for layer in layers_with_gap:
    box = FancyBboxPatch(
        (layer["x"], layer["y"]), layer["width"], layer["height"],
        boxstyle="round,pad=0.1", edgecolor="black", facecolor="white"
    )
    ax.add_patch(box)
    # Add text
    plt.text(
        layer["x"] + layer["width"] / 2, layer["y"] + layer["height"] / 2,
        layer["name"], ha="center", va="center", fontsize=10
    )

# Add arrows between layers
for i in range(len(layers_with_gap) - 1):
    x_start = layers_with_gap[i]["x"] + layers_with_gap[i]["width"]
    x_end = layers_with_gap[i + 1]["x"]
    y = layers_with_gap[i]["y"] + layers_with_gap[i]["height"] / 2
    plt.arrow(x_start, y, x_end - x_start - 0.1, 0, head_width=0.1, head_length=0.2, fc='black', ec='black')

# Configure plot limits and appearance
plt.xlim(-1, 12)
plt.ylim(-1, 1)
plt.axis('off')  # Turn off the axis

# Display the plot
plt.title("Model RNN", fontsize=12)
plt.show()
```

```python

from sklearn.metrics import classification_report

report = classification_report(y_true_classes, y_pred_classes, output_dict=True)
classes = list(report.keys())[:-3]  # Exclude 'accuracy', 'macro avg', 'weighted avg'

precision = [report[cls]['precision'] for cls in classes]
recall = [report[cls]['recall'] for cls in classes]
f1_scores = [report[cls]['f1-score'] for cls in classes]

# Bar plot
x = np.arange(len(classes))
plt.figure(figsize=(10, 6))
plt.bar(x - 0.2, precision, width=0.2, label='Precision')
plt.bar(x, recall, width=0.2, label='Recall')
plt.bar(x + 0.2, f1_scores, width=0.2, label='F1-Score')
plt.xticks(x, classes, rotation=45)
plt.title('Class-wise Metrics')
plt.xlabel('Classes')
plt.ylabel('Scores')
plt.legend()
plt.show()
```
